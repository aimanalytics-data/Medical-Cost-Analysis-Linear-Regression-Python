# lets evaluate our model

* using Training and Testing method
we will train or teach our machine learning model with % of our data and test it with % of our data set to evaluate the performance of the model

```python
x = df[["age", "bmi", "smoker_encoded"]]
y = df["medical charges_winsorized"]
```

```python
x_data=df.drop('medical charges_winsorized',axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
model=LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
```

```python
print("MSE on Training Set:", mse_train)
print("MSE on Testing Set:", mse_test)
```

The Mean Squared Error for the training and testing se is very close. This close MSE between training and testing data indicates that the model is not overfitting, as it performs almost as well on unseen data as it does on the training data.
